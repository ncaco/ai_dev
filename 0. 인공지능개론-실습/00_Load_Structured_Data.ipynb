```python
## 준비
path1 = '/content/D1-Exp-and-Salary.txt'
path2 = '/content/D1-Unknown-but-for-Regression.txt'
path3 = '/content/D1-Pass-or-Fail.txt'
```

```python
## 1) 전통적 C 언어 스타일의 파일 처리

# 데이터를 저장할 빈 리스트를 초기화
X = []
y = []

# 파일을 열고 각 줄을 반복 처리
with open(path1, 'r') as file:
    for line in file:
        # 줄바꿈 문자를 제거하고, 쉼표로 각 값을 분리
        values = line.strip().split(',')

        # 마지막 값을 제외한 모든 값을 X에 추가. 값들을 float 형태로 변환
        X.append([float(value) for value in values[:-1]])

        # 마지막 값을 y에 추가. 값을 float 형태로 변환
        y.append(float(values[-1]))

# 결과를 출력해 확인
print("X:", X)
print("y:", y)
```

```python
## 2) pandas를 이용한 처리

import pandas as pd

# pandas를 사용하여 txt 파일을 읽습니다. header=None 옵션은 칼럼 헤더가 없음을 의미합니다.
df = pd.read_csv(path2, header=None, sep=',')

# 독립 변수 X: 마지막 칼럼을 제외한 모든 칼럼
X = df.iloc[:, :-1].values

# 종속 변수 y: 마지막 칼럼
y = df.iloc[:, -1].values

# 결과를 출력해 확인합니다.
print("X:", X)
print("y:", y)
```

```python
## 3) numpy의 genfromtxt를 이용한 처리

import numpy as np

# numpy의 genfromtxt 함수를 사용하여 데이터를 불러옵니다.
data = np.genfromtxt(path3, delimiter=',')

# 독립 변수 X: 모든 행과 마지막 열을 제외한 모든 열
X = data[:, :-1]

# 종속 변수 y: 모든 행의 마지막 열
y = data[:, -1]

# 결과를 출력해 확인합니다.
print("X:", X)
print("y:", y)
```

```python
## 4) 행렬 데이터 연산 해보기
import numpy as np

# 1. 평균 구하기
X_mean = np.mean(X, axis=0)  # X의 각 열별 평균
y_mean = np.mean(y)          # y의 평균
print("X 평균:", X_mean)
print("y 평균:", y_mean)

# 2. 표준편차 구하기
X_std = np.std(X, axis=0)  # X의 각 열별 표준편차
y_std = np.std(y)          # y의 표준편차
print("X 표준편차:", X_std)
print("y 표준편차:", y_std)

# 3. 최댓값과 최솟값 구하기
X_max, X_min = np.max(X), np.min(X)
y_max, y_min = np.max(y), np.min(y)
print("X 최댓값:", X_max, "X 최솟값:", X_min)
print("y 최댓값:", y_max, "y 최솟값:", y_min)

# 4. 데이터 정규화 (Min-Max Scaling)
X_mm_normalized = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
print("Min-Max 정규화된 X:\n", X_mm_normalized)

# 5. 표준 정규화 (Normalization)
X_zscore_normalized = (X - X.mean(axis=0)) / X.std(axis=0)
# 정규화된 데이터를 출력합니다.
print("Normalized X:\n", X_zscore_normalized)
```
